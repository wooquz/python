```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
plt.plot(x, y, label = 'sin(x)',color = 'red')
plt.plot(x, z, label = 'cos(x)', color = 'yellow')
plt.title('Regular graphics')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.random.normal(0, 1, 1000)
plt.hist(x, bins = 20, color='red')
plt.title('')
plt.xlabel('')
plt.ylabel('')
plt.show()
```
